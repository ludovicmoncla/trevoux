# Load TSV into DataFrame and explore the data


```python
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join('..', 'data', '1743_LeRobert', 'Trevoux1743.tsv'), sep='\t')
df.head(10)
```

```python
df.shape
```

```python
# Nombre de définitions (<sense>) par volume
df['volume'].value_counts()
```

```python


df['subordinate_domain_norm'] = df['subordinate_domain'].fillna('')
df['subordinate_domain_norm'] = df['subordinate_domain_norm'].str.lower()
df['subordinate_domain_norm'] = df['subordinate_domain_norm'].str.replace('termes', 'terme')
df['subordinate_domain_norm'] = df['subordinate_domain_norm'].str.replace('terme de ', '')
df['subordinate_domain_norm'] = df['subordinate_domain_norm'].str.replace("terme d'", '')
df['subordinate_domain_norm'] = df['subordinate_domain_norm'].str.replace("de ", '')
df['subordinate_domain_norm'] = df['subordinate_domain_norm'].str.replace("d'", '')

# domaines triés par fréquence
df['subordinate_domain_norm'].value_counts().head(15)
```
