## File per volume

```python
import os
import re
import pandas as pd

def escape_selective(text):
    # On cherche soit une balise, soit un caractère spécial
    pattern = r'(<[^>]+>)|([&<>])'
    
    def subst(match):
        balise = match.group(1)
        char = match.group(2)
        
        if balise: return balise # On rend la balise telle quelle
        if char == '&': return '&amp;'
        if char == '<': return '&lt;'
        if char == '>': return '&gt;'
        
    return re.sub(pattern, subst, text)

volume_gallica = {1: 'bpt6k8542165', 2: 'bpt6k1506801d', 3: 'bpt6k15068000', 4: 'bpt6k15068037', 5: 'bpt6k1913247n', 6: 'bpt6k1506802t'}
page_gallica = {1: '40', 2: '12', 3: '12', 4: '12', 5: '12', 6: '12'}

# 1. Fonction pour générer uniquement le bloc XML d'une entrée
def get_entry_xml(entry, id, volume, head, src_domain,  text, macro_domain, pred_domain):
    #TODO : probleme du replace pour les balises <i> ou <br> qui sont aussi remplacées. A revoir.
    #text = text.replace('&', '&amp;').replace('<', '&lt;').replace('>', '&gt;')  # Échapper les caractères spéciaux
    text = escape_selective(text)

    text = text.replace('&amp;#9758;', '&#9758;') # manicule
    
    # replace newlines with <br/> for better formatting in XML
    text = text.replace('\n\n', '<br/>')
    text = text.replace('\n', '<br/>')
    
    manicule = True if '&#9758;' in text else False
       
    # add url for "voyez" 
    #"Voyez" followed by one word and a dot. Ex: Voyez ACCARER. devient Voyez <a href="http://trevoux.huma-num.fr/philologic/DUFLT/bibliography?&head=ACCARER">ACCARER</a>
    text = re.sub(r'Voyez\s+(\w+)\.', r'Voyez <ref type="search" target="head:\1">\1</ref>.', text)

    # turn <pb n="2"/> into <pb n="2" facs="bpt6k8542165/f42"/>
    #text = re.sub(r'<pb n="(\d+)"\s*/>', r'<pb n="\1" facs="' + volume_gallica[volume] + '/f'+page_gallica[volume]+\1+'"/>', text)
    text = re.sub(
        r'<pb n="(\d+)"\s*/>',
        lambda m: f'<pb n="{m.group(1)}" facs="{volume_gallica[volume]}/f{int(page_gallica[volume]) + int(m.group(1))}/full/1200,/0/native.jpg"/>',
        text
    )

    return f'''
<div vol="{volume}" n="{entry}" id="{id}" type="{src_domain}">
   <index type="head" value="{head}" />
   <index type="src_domain" value="{src_domain}" />
   <index type="macro_domain" value="{macro_domain}" />
   <index type="pred_domain" value="{pred_domain}" />
   <index type="manicule" value="{str(manicule).lower()}" />
    <head>{head}</head>
    <p>
    {text}
    </p>
</div>'''

# 2. Fonction pour générer l'enveloppe globale du volume
def get_volume_xml(book, volume, body_content):
    return f'''<TEI xmlns="http://www.tei-c.org/ns/1.0">
<teiHeader>
   <fileDesc>
      <titleStmt>
         <title>Volume {volume}</title>
      </titleStmt>
      <publicationStmt>
         <distributor>
            <name></name>
         </distributor>
      </publicationStmt>
      <sourceDesc>
         <bibl>
            <creation>
                  <date>{book}</date>
            </creation>
         </bibl>
      </sourceDesc>
   </fileDesc>
</teiHeader>
<text>
<body>
{body_content}
</body>
</text>
</TEI>
'''

# Chemin de sortie (à adapter si nécessaire)
output_dir = os.path.join('..', 'data', '1743_LeRobert', 'Philologic', 'DUFLT')
os.makedirs(output_dir, exist_ok=True) # Crée le dossier s'il n'existe pas

# 3. Boucle principale groupée par volume
# On suppose que votre DataFrame s'appelle 'df_sample'
# On groupe les données par la colonne 'volume'
for volume_name, group_df in df_sample.groupby('volume'):
    
    entries_xml_list = []
    
    # On récupère la date (book) de la première entrée du volume pour le Header
    # (On suppose que la date est la même pour tout le volume)
    current_book_date = group_df.iloc[0]['book']
    
    # On itère sur chaque entrée DANS ce volume spécifique
    for index, row in group_df.iterrows():
        duflt_domain = row['srcDomain'] if pd.notnull(row['srcDomain']) else ''
        macro_domain = row['macroDomain'] if pd.notnull(row['macroDomain']) else ''
        pred_domain = row['predDomain'] if pd.notnull(row['predDomain']) else ''
        
        # On génère le petit bloc div pour cette entrée
        entry_block = get_entry_xml(
            row['numero'], 
            row['subEntryId'],
            row['volume'], 
            row['head'], 
            duflt_domain, 
            row['html'],
            macro_domain,
            pred_domain
        )
        entries_xml_list.append(entry_block)

    # On joint toutes les entrées en une seule chaîne de caractères
    full_body_content = "\n".join(entries_xml_list)
    
    # On insère ce contenu dans l'enveloppe du volume
    final_tei_xml = get_volume_xml(current_book_date, volume_name, full_body_content)

    # 4. Écriture du fichier (Nommé par volume cette fois)
    # Nettoyage simple du nom de fichier pour éviter les caractères spéciaux si nécessaire
    #safe_vol_name = str(volume_name).replace(" ", "_")
    output_filename = f"volume_{volume_name}.xml"
    
    output_path = os.path.join(output_dir, output_filename)
    
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(final_tei_xml)

    print(f"Fichier généré : {output_filename} ({len(group_df)} entrées)")
```
